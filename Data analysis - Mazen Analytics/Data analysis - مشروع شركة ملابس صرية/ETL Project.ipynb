## Full ETL

```python
import os
from natsort import natsorted

dfs = []

files = os.listdir('Data/')
sorted_files = natsorted(files)   # Sort files
sorted_files
```

### Method1:

```python
for file in files:
  # Extract
  df = pd.read_html('Data/' + file)
  df = df[0]

  if len(dfs) > 0:
    df.columns = ['القيمة', 'الكمية', 'السعر', 'الصنف', 'الكود', 'الباركود', 'رقم الإذن', 'التاريخ']
    if len(dfs) == len(files)-1:
      df.drop(index=29, axis=0, inplace=True)
      
  # Transform
  dfs.append(df)
  print(df.shape)

clothes_df = pd.concat(dfs)  
clothes_df.drop(columns=['الكود'], inplace=True)
clothes_df.reset_index(drop=True, inplace=True)
clothes_df
```

```python
# Load
# We don't need index because Excel has it already
clothes_df.to_excel('clothes.xlsx', index=False)    
```
